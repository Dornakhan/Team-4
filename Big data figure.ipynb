```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create figure with specific layout for PowerPoint
fig = plt.figure(figsize=(14, 8))

# Title section
fig.text(0.5, 0.95, 'Top 2: Anti-Social Behaviour (ASB)', fontsize=20, weight='bold', ha='center')
fig.text(0.5, 0.91, 'Loud noise • Groups intimidating • Vandalism • Drunk & disorderly', 
         fontsize=12, ha='center', style='italic', color='gray')

# Create main comparison panel
ax = plt.subplot(2, 1, 1)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
asb_2023 = [1500, 1700, 1900, 2200, 2300, 2400, 2200, 2150, 2000, 1900, 1650, 1450]
asb_2024 = [1500, 1700, 1900, 2200, 2300, 2400, 2400, 2000, 2400, 1900, 2100, 1500]

# Plot both years
ax.plot(months, asb_2023, 'g-o', linewidth=3, markersize=10, label='2023 - Normal Pattern')
ax.plot(months, asb_2024, 'r-s', linewidth=3, markersize=10, label='2024 - Disrupted Pattern')

# Mark the riots
ax.axvline(x=7, color='black', linestyle='--', linewidth=2)
ax.text(7, 2500, 'AUGUST 3rd\nRIOTS', ha='center', fontsize=11, weight='bold',
        bbox=dict(boxstyle="round,pad=0.3", facecolor='yellow', edgecolor='black'))

# Highlight the chaos period
ax.axvspan(7, 11, alpha=0.2, color='red')
ax.text(9, 2550, 'CHAOTIC PERIOD', ha='center', fontsize=12, weight='bold', color='red')

# Add arrows showing up-down pattern
for i in range(8, 11):
    if i % 2 == 0:
        ax.annotate('', xy=(i, asb_2024[i]), xytext=(i, asb_2024[i]-150),
                   arrowprops=dict(arrowstyle='->', color='red', lw=2))
    else:
        ax.annotate('', xy=(i, asb_2024[i]), xytext=(i, asb_2024[i]+150),
                   arrowprops=dict(arrowstyle='->', color='green', lw=2))

ax.set_ylabel('Number of Incidents', fontsize=12)
ax.set_title('ASB Patterns: 2023 vs 2024', fontsize=14, weight='bold')
ax.legend(loc='upper left', fontsize=11)
ax.grid(True, alpha=0.3)
ax.set_ylim(1200, 2700)

```

```python
# ASB-specific figure showing 2023 vs 2024 patterns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 2023 Pattern
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
asb_2023 = [1500, 1700, 1900, 2200, 2300, 2400, 2200, 2150, 2000, 1900, 1650, 1450]

ax1.plot(months, asb_2023, 'g-o', linewidth=3, markersize=8)
ax1.fill_between(range(5, 8), 1400, 2500, alpha=0.3, color='yellow', label='Summer Peak')
ax1.set_title('2023: Normal Seasonal Pattern', fontsize=14, weight='bold')
ax1.set_ylabel('Incidents')
ax1.set_ylim(1200, 2600)
ax1.grid(True, alpha=0.3)
ax1.text(6.5, 1300, 'Weather-driven', ha='center', fontsize=11, style='italic')
ax1.legend()

# 2024 Pattern
asb_2024 = [1500, 1700, 1900, 2200, 2300, 2400, 2400, 2000, 2400, 1900, 2100, 1500]

ax2.plot(months, asb_2024, 'r-o', linewidth=3, markersize=8)
ax2.fill_between(range(7, 12), 1400, 2500, alpha=0.3, color='red', label='Chaos Period')
ax2.axvline(x=7, color='black', linestyle='--', linewidth=2)
ax2.text(7.1, 2300, 'RIOTS', rotation=90, fontsize=12, weight='bold')

# Add up-down arrows after riots
for i in range(8, 11):
    if i % 2 == 0:
        ax2.arrow(i, 2100, 0, 200, head_width=0.2, head_length=50, fc='green', ec='green')
    else:
        ax2.arrow(i, 2200, 0, -200, head_width=0.2, head_length=50, fc='red', ec='red')

ax2.set_title('2024: Post-Riot Chaos (Police Patrols Vary)', fontsize=14, weight='bold')
ax2.set_ylabel('Incidents')
ax2.set_ylim(1200, 2600)
ax2.grid(True, alpha=0.3)
ax2.legend()

plt.suptitle('ANTI-SOCIAL BEHAVIOUR: From Predictable to Chaotic', fontsize=16, weight='bold')
plt.tight_layout()
plt.show()
```
